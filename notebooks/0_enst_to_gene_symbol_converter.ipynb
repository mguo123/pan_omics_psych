# ENST to gene symbol converter

```python
import mygene
mg = mygene.MyGeneInfo()
import pandas as pd
import os,glob
```

```python
enst_ids = pd.read_table('../data/raw/rna/HEK293T/SRR3997504.tsv')
enst_ids['enst'] = enst_ids.target_id.apply(lambda x: x.split('.')[0])
```

```python
enst_ids['enst'].unique().shape,enst_ids['enst'].shape
```

```python
%%time
enst_sym_df = pd.DataFrame(mg.querymany(enst_ids['enst'], scopes='ensembl.transcript', species='human'))
```

```python
enst_sym_df.to_csv('../data/external/enst_sym_df.csv')
```
